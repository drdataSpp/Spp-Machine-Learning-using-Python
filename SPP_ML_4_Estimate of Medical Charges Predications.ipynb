# Pre-Processing the data and Partitioning it

```python
medical_insurance_df.columns
```

```python
#encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
categorical_variables = ['sex','smoker', 'region']

for columns in categorical_variables:
    medical_insurance_df[columns] = le.fit_transform(medical_insurance_df[columns])
```

```python
medical_insurance_df.head()
```

```python
#Spliting the whole data set into training and test sets
from sklearn.model_selection import train_test_split
```

```python
X = medical_insurance_df.drop("charges", axis=1)
```

```python
y = medical_insurance_df["charges"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Training several models and choosing the best one!

## 1) Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor
dTree = DecisionTreeRegressor()
```

```python
dTree.fit(X_train, y_train)
```

```python
y_pred_dTree = dTree.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print("R-Squared Score of the Decision Tree model is", round(r2_score(y_test, y_pred_dTree) * 100, 2))
```

```python
print("Root mean squared error of the Decision Tree model is ", np.sqrt(mean_squared_error(y_test, y_pred_dTree)))
```

## 2) Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=1000)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred_rfr = rfr.predict(X_test)
```

```python
print("R-Squared Score of the Random Forest Classifier is", round(r2_score(y_test, y_pred_rfr) * 100, 2))
```

```python
print("Root mean squared error of the Random Forest Classifier is ", np.sqrt(mean_squared_error(y_test, y_pred_rfr)))
```
